```python
import requests
from pprint import pprint

# Step 1 - create
response = requests.post(
    "http://localhost:8002/api/events/",
    json={"page": "home", "description": "inline test"}
)
response.json()
```

```python
# Step 2 - read all
response = requests.get("http://localhost:8002/api/events/")
pprint(response.json())  # should include the event you just created
```

```python
import random

event_count = 1000

pages = ['/about', '/contact', '/pages', 'pricing']

for i in range(event_count):
    page = random.choice(pages)
    response = requests.post(
        "http://localhost:8002/api/events/",
                             json={"page": page})

    if response.ok:
        data = response.json()
        print(data, type(data), data.get('items'))
    else:
        print(response.text)
```
