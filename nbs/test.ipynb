```python
import requests
from pprint import pprint

# Step 1 - create
response = requests.post(
    "http://localhost:8002/api/events/",
    json={"page": "home", "description": "inline test"}
)
response.json()
```

```python
# Step 2 - read all
response = requests.get("http://localhost:8002/api/events/")
pprint(response.json())  # should include the event you just created
```
